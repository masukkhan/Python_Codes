```python
import string

plaintext=input("enter the plain text to encrypt \n")

key=input("enter the key for encryption \n")

cipher=""
for c in plaintext:
    if c in string.ascii_lowercase:
        index=ord(c)-ord('a')
        cipher=cipher+key[index]
    else:
        cipher=cipher+c
print("\n plaintext :"+plaintext+"\n")
print("ciphertext :"+ cipher+"\n")
```
