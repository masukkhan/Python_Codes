```python
import random 
from art import logo 
print(logo)


easy_life = 10 
hard_life = 5

def check_answer(computer_guess,your_guess,life):
  if computer_guess > your_guess:
    print("Too low")
    return life-1
  elif computer_guess < your_guess:
    print("Too high")
    return life -1 
  else:
    print("You got it . Its the correct answer")
    return life -1
def difficulty():
  level = input("which level you want to play 'easy' or 'hard' ")
  if level == 'easy':
    return easy_life
  elif level == "hard":
    return hard_life

def game():
  print("Welcome to the Number Guessing Game!")
  print("I'm thinking of a number between 1 and 100.")
  computer_guess = random.randint(0,100)
  print(computer_guess)
  life = difficulty()
  while life !=0:
    
    print(f"You Have {life} chances to guess the number ")
    your_guess = int(input("Now You guess a number "))
    life = check_answer(computer_guess,your_guess,life)
    if life !=0 :
      print("Guess Again ")
    else:
      print("You lost all your life play again ")

game()
```
